#clean the null values

- they are 3 types
-1)dropping missing values(dropna()):
  - it is the way to deal with missing data to drop it
  - dropna() is used for this method
  - we can using this method entire null data will loss  

```python
import pandas as pd
data = pd.read_csv(r"/content/sample_data/Diabetes Missing Data.csv")
data.head(10)
```

```python
#finding null values
data.isnull().sum()
```

```python
#shape
print(data.shape)
#using dropna method
data1 = data.dropna()
print(data1.isnull().sum())
data1.shape
```

```python
#using inplace method
import pandas as pd
data = pd.read_csv(r"/content/sample_data/Diabetes Missing Data.csv")
data.head()
print(data.shape)
data.dropna(inplace=True,axis = 0)
print(data.isnull().sum())
print(data.shape)
```

2)replacing missing values:

- we can replace the values with mean or median
- we can call it as impitation or replacement technique

```python
import pandas as pd
data = pd.read_csv("/content/sample_data/Diabetes Missing Data.csv")
print(data.isnull().sum())
data.shape
```

```python
#replace null values using mean
mean_value = data.mean()
print(mean_value)
data1 = data.fillna(mean_value)
data1.head()
```

```python
#using median
median_value = data.median()
data2 = data.fillna(median_value)
print(median_value)
print(data2.shape)
data2.head()
```


3)duplicate data:

- it will drop the repeating values
- it will checks row by row if there is any repeated values it can drop

```python
import pandas as pd
data = pd.read_csv("/content/sample_data/Diabetes Missing Data.csv")
print(data.isnull().sum())
print(data.shape)
data1 = data.drop_duplicates()
data1.shape
```

#aggregating data with groupby() in pandas:

- it is used for summarizing the data aggregating function more than 2 columns
- aggregative known as sum and average
- when we apply groupby method we need to clean the data

```python
import pandas as pd
data = pd.read_csv("/content/sample_data/Diabetes Missing Data.csv")
data.head()
#clean the data
median_value = data.median()
print(median_value)
data1 = data.fillna(median_value).head(20)
data1.head(20)
data1.isnull().sum()
data1.groupby("Glucose").median().head()
```
